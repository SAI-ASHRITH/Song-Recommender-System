```python
adf=adf.sort_values('Track')    #Sorting the rows of artist data in ascending order of the column 'Track'.
```

```python
adf=adf.drop_duplicates()   #Drop the duplicate and repeating values.
```

```python
adf         #Print the artist data to check for duplicates.
```

```python
adf.count()         #Total count of unique artists. We have 185 unique artist rows.
```

```python
adf_dict=dict(zip(adf['Track'],adf['Artist']))    #Create a dictionary with 'Track' as keys and 'Artist' as values.
```

```python
adf_dict      #Print the dictionary.
```

```python
x=[]                                      #Create an empty list.
for i in tdf['Track']:
    x.append(adf_dict[i])                 #For all the ids in 'Track', append the corresponding 'Artist', from the created dictionary, to the list.
tdf.insert(2, "Artist", x, True)          #Insert the created list as a cloumn in the traininf dataframe.
```

```python
tdf         #Print the new training data to check for successfull insertion.
```

```python
tdf.to_csv('newtrain.csv',index=False)      #Save the new training data into a file.
```
